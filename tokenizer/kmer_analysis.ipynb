```python
import re

def remove_singleton_numbers(smiles):
    # Regex to match all numbers (including %10 for multi-digit cyclizations)
    number_pattern = r'%\d+|\[\d+|\d'
    
    # Find all numbers in the SMILES string
    numbers = re.findall(number_pattern, smiles)
    
    # Count occurrences of each number
    number_counts = {num: numbers.count(num) for num in set(numbers)}
    
    # Identify numbers that appear exactly once
    singleton_numbers = {num for num, count in number_counts.items() if count == 1}

    # Result string
    result = []
    i = 0
    in_brackets = False
    
    while i < len(smiles):
        char = smiles[i]
        
        if char == '[':
            # Entering a bracketed section
            in_brackets = True
            result.append(char)
            i += 1
            continue
        elif char == ']':
            # Exiting a bracketed section
            in_brackets = False
            result.append(char)
            i += 1
            continue

        if in_brackets:
            # Inside a bracketed section, preserve all characters
            result.append(char)
            i += 1
            continue
        
        # Check for multi-digit cyclization
        if char == '%' and i + 2 < len(smiles) and smiles[i+1:i+3].isdigit():
            num = smiles[i:i+3]  # Extract %XX
            if num not in singleton_numbers:
                result.append(num)
            i += 3
        elif char.isdigit():  # Check for single-digit numbers
            num = char
            if num not in singleton_numbers:
                result.append(num)
            i += 1
        else:
            # Non-number characters
            result.append(char)
            i += 1
    
    return ''.join(result)

# Test cases
smiles_examples = [
    # "C1CC1",               # Nothing removed (1 appears twice)
    # "C1CC",                # Remove 1 (appears once)
    # "C1CC2CCC2",           # Nothing removed (1 and 2 appear twice)
    # "[13C]1CC1",           # Nothing removed (1 appears twice; [13C] is preserved)
    # "C1CCC2CCC3CCCC3C2C1", # Nothing removed (all numbers appear multiple times)
    # "C1CC%10CCCC%10",      # Nothing removed (%10 appears twice)
    # "C1CC1C2CCC3CCCC3",    # Remove 2 (appears once; keep others)
    "[12C][13C]1CC",       # Remove 1 (appears once; [12C] and [13C] preserved)
    "N1[C@@H](CCC1)C(=O)N[C@H](CC(=O)O)C(=O)N[C@@H](CC1=CNC=N1)C(=O)N[C@@H](CO)C(=O)N[C@@H]([C@H](CC)C)C(=O)N[C@@H](Cc1ccccc1)C(=O)N[C@@H](CC(C)C)C(=O)N[C@@H](CC(=O)N3)C(=O)N[C@@H](CC(C)C)C(=O)N[C@@H](CO)C(=O)N[C@@H](Cc1ccccc1)C(=O)NCC(=O)N[C@H](CCC[C@H](C(=O)O)N)C(=O)N[C@@H](CC(=O)N)C(=O)N[C@H](CCC[C@H](C(=O)O)N)C(=O)N[C@@H](CC(=CN2)C1=C2C=CC=C1)C(=O)N[C@@H](CCSC)C(=O)N[C@@H](CCSC)C(=O)N[C@@H]([C@H](O)C)C(=O)N[C@@H](CC(C)C)C(=O)N[C@@H](CO)C(=O)N[C@@H](C)C(=O)N[C@@H](CCC(=O)O)C(=O)N[C@@H](CCCNC(=N)N)C(=O)N[C@@H](C(C)C)C(=O)N[C@@H](CC(=O)N)C(=O)N[C@@H](CCCNC(=N)N)C(=O)N[C@@H](CC(=O)O)C(=O)N[C@@H](Cc1ccccc1)C(=O)N[C@@H](CC1=CNC=N1)C(=O)N[C@@H](CCCNC(=N)N)C(=O)N[C@@H](CCSC)C(=O)N[C@@H](CC1=CNC=N1)C(=O)N[C@@H](C(C)C)C(=O)N[C@@H]([C@H](O)C(=O)O)C(=O)N[C@@H](CC1=CNC=N1)C(=O)N[C@@H](CCC(=O)O)C(=O)N[C@@H](C)C(=O)N[C@H](CS)C(=O)N[C@@H](Cc1ccccc1Cl)C(=O)N[C@@H](CCC(=O)N)C(=O)N[C@@H](CCSC)C(=O)N[C@@H](CCC(=O)N)C(=O)N[C@@H](CCCNC(=N)N)C(=O)N[C@@H](Cc1c[nH]c2c1c(N)ccc2)C(=O)O"
]

for smiles in smiles_examples:
    print(f"Original: {smiles} -> Processed: {remove_singleton_numbers(smiles)}")
```

```python
from rdkit import Chem
from tqdm import tqdm
import pandas as pd
import numpy as np
from rdkit import RDLogger

RDLogger.DisableLog('rdApp.error')

smiles = pd.read_csv('all_smiles_fixed(parenth_dash_loops).txt', header=None)
smiles_list = np.array(smiles).flatten().tolist()
print(len(smiles_list))

canonical_smiles = []
none_errors = 0
processing_errors = 0

for smile in tqdm(smiles_list):
    try:
        mol = Chem.MolFromSmiles(smile)
        if mol is None:
            none_errors += 1
            continue  # Skip this SMILES
        canonical_smiles.append(Chem.MolToSmiles(mol))
    except Exception as e:
        processing_errors += 1
        continue  # Skip this SMILES

# save canonical smiles
with open('400k_canonical_smiles.txt', 'w') as f:
    for item in canonical_smiles:
        f.write("%s\n" % item)

# Print error summary
print(f"Total SMILES processed: {len(smiles)}")
print(f"SMILES with MolFromSmiles=None: {none_errors}")
print(f"SMILES with MolToSmiles errors: {processing_errors}")
print(f"Successfully processed SMILES: {len(canonical_smiles)}")

# Total SMILES processed: 400000
# SMILES with MolFromSmiles=None: 18034
# SMILES with MolToSmiles errors: 0
# Successfully processed SMILES: 381966
```
